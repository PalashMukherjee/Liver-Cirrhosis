```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Age

# Create a figure with 2 subplots in a single row
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Customize style
sns.set_style("ticks")
sns.set_palette("pastel")

# Plot 1: Density plot
sns.kdeplot(df['Age'], ax=axs[0], shade=True)
axs[0].set_title('Density plot of Age', fontsize=16)
axs[0].set_xlabel('Age', fontsize=12)
axs[0].set_ylabel('Density', fontsize=12)

# Plot 2: Box plot
sns.boxplot(x='Classes', y='Age', data=df, ax=axs[1])
axs[1].set_title('Boxplot of Age for Diseased or not Diseased', fontsize=16)
axs[1].set_xlabel('Class', fontsize=12)
axs[1].set_ylabel('Age', fontsize=12)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
from scipy.stats import shapiro

# Assuming 'df' is your pandas DataFrame containing the data
# Grouping the DataFrame by 'Classes' column
grouped = df.groupby('Classes')

# Creating an empty list to store the results
results = []

# Looping through each group
for name, group in grouped:
    for column in group.drop(['gender', 'Classes'], axis=1):
        stat, p = shapiro(group[column])
        alpha = 0.05
        if p > alpha:
            normality = 'Gaussian (fail to reject H0)'
        else:
            normality = 'Not Gaussian (reject H0)'
        results.append({'Variable': column, 'Class': name, 'Statistics': stat, 'p-value': p, 'Normality': normality})

# Creating a DataFrame from the results list
results_df = pd.DataFrame(results)

# Displaying the results
print(results_df)
```
